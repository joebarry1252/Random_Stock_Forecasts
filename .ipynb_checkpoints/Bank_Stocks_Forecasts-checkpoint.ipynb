```python
import numpy as np
import pandas as pd
import os
import alpaca_trade_api as tradeapi
import datetime as dt
import pytz
from MCForecastTools import MCSimulation
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
# Set the ticker
bank_tickers = ["BAC","GS","JPM","MS","WFC"]

# Set timeframe to '1D'
timeframe = "1D"

# Set start and end datetimes of 5 years from Today
start_date = pd.Timestamp("2016-06-06", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-06-06", tz="America/New_York").isoformat()

# Get 5 years worth of historical data for banks
bank_ticker_data = api.get_barset(
    bank_tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df

bank_ticker_data.head()
```

```python
num_sims = 1000

# Configure a Monte Carlo simulation to forecast three years daily returns
MC_banks = MCSimulation(
    portfolio_data = bank_ticker_data,
    num_simulation = num_sims,
    num_trading_days = 252*5
)

MC_banks_even = MCSimulation(
    portfolio_data = bank_ticker_data,
    weights = [.2, .2, .2, .2, .2],
    num_simulation = num_sims,
    num_trading_days = 252*5
)
```

```python
MC_banks.calc_cumulative_return()
```

```python
banks_simulated_returns_data = {
    "mean": list(MC_banks.simulated_return.mean(axis=1)),
    "median": list(MC_banks.simulated_return.median(axis=1)),
    "min": list(MC_banks.simulated_return.min(axis=1)),
    "max": list(MC_banks.simulated_return.max(axis=1))
}

# Create a DataFrame with the summary statistics
banks_simulated_returns = pd.DataFrame(banks_simulated_returns_data)

# Display sample data
banks_simulated_returns.head()
```

```python
banks_simulated_returns.plot(title="Simulated Daily Returns Behavior of Banks Stock Over Next Year")
```

```python
initial_investment = 10000

banks_cum_pnl = initial_investment * banks_simulated_returns
banks_cum_pnl.head()
```

```python
banks_cum_pnl.plot()
```

```python
MC_banks_even.calc_cumulative_return()
```

```python
# Fetch summary statistics from the Monte Carlo simulation results
bank_even_tbl = MC_banks_even.summarize_cumulative_return()

# Print summary statistics
print(bank_even_tbl)
```

```python
even_ci_lower = round(bank_even_tbl[8]*10000,2)
even_ci_upper = round(bank_even_tbl[9]*10000,2)

# Print results
print(f"There is a 95% chance that an initial investment of $10,000 in the portfolio"
      f" over the next 5 years will end within in the range of"
      f" ${even_ci_lower} and ${even_ci_upper}.")
```
