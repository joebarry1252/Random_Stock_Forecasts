```python
import os
import pandas as pd
import numpy as np
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
%matplotlib inline
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
# 5 Years worth of data

# Set timeframe to '1D'
timeframe = "1D"


start_date = pd.Timestamp("2016-06-23", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-06-23", tz="America/New_York").isoformat()

# Set the ticker information
tickers = ["AAPL","AMZN","FB","GOOGL"]


df_ticker = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df

# Display sample data
df_ticker.head()
```

```python
num_sims = 1000

MC_FAAG = MCSimulation(
    portfolio_data = df_ticker,
    num_simulation = num_sims,
    num_trading_days = 1260
)
```

```python
MC_FAAG.calc_cumulative_return()
```

```python
FAAG_lineplot = MC_FAAG.plot_simulation()
```

```python
simulated_returns_data = {
    "mean": list(MC_FAAG.simulated_return.mean(axis=1)),
    "median": list(MC_FAAG.simulated_return.median(axis=1)),
    "min": list(MC_FAAG.simulated_return.min(axis=1)),
    "max": list(MC_FAAG.simulated_return.max(axis=1))
}


df_simulated_returns = pd.DataFrame(simulated_returns_data)


df_simulated_returns.head()
```

```python
df_simulated_returns.plot(title="Simulated Daily Returns Behavior of FB, AAPL, AMZN, and GOOGL Stock Over the Next 5 Years")
```

```python
tbl = MC_FAAG.summarize_cumulative_return()

ci_lower = round(tbl[8]*10000,2)
ci_upper = round(tbl[9]*10000,2)

print(f"There is a 95% chance that an initial investment of $10,000 in the portfolio"
      f" over the next five years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}.")
```
